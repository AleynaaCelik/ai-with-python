```python
from matplotlib import pyplot as plt  # Alias: plt
import numpy as np                    # Alias: np
```

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 3*np.pi, 1000)
ax.plot(x, np.sin(x))

# ax.set(xlabel='x', ylabel='sin(x)', title='A simple plot')

ax.set_xlabel('x')
ax.set_ylabel('sin(x)')
ax.set_title('A simple plot')

fig.savefig("test.png")
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 3*np.pi, 1000)
ax.plot(x, np.sin(x), label = 'sin(x)')
ax.plot(x, np.cos(x), label = 'cos(x)')

ax.set(xlabel='x', ylabel='f(x)', title='A multiple plot')

ax.legend(loc='lower left')
ax.grid()

fig.savefig("multiple_test.png")
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 3*np.pi, 1000)
ax.plot(x, np.sin(x), label = 'sin(x)', color = 'red')
ax.plot(x, np.cos(x), label = 'cos(x)', color = 'blue')
ax.plot(x, 2*np.sin(x) + 3*np.cos(4*x), label = 'Linear Comb.', color = 'green')

ax.set(xlabel='x', ylabel='f(x)', title='A multiple plot')

ax.legend(loc='lower left')
ax.grid()

fig.savefig("multiple_test.png")
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 3*np.pi, 1000)
ax.plot(x, np.sin(x), label = 'sin(x)', color = 'red', linestyle = 'dashed')
ax.plot(x, np.cos(x), label = 'cos(x)', color = 'blue', linestyle = 'dotted')
ax.plot(x, 2*np.sin(x) + 3*np.cos(4*x), label = 'Linear Comb.', color = 'green', linestyle = 'solid')
ax.plot(x, np.sin(6*x), label = 'sin(6x)', color = 'purple', linestyle = 'dashdot')

ax.set(xlabel='x', ylabel='f(x)', title='A multiple plot')

ax.legend(loc='lower left')
ax.grid()

fig.savefig("multiple_test.png")
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 3*np.pi, 1000)
ax.plot(x, np.sin(x), label = 'sin(x)', color = 'red', linestyle = 'dashed')
ax.plot(x, np.cos(x), label = 'cos(x)', color = 'blue', linestyle = 'dotted')
ax.plot(x, 2*np.sin(x) + 3*np.cos(4*x), label = 'Linear Comb.', color = 'green', linestyle = 'solid')
ax.plot(x, np.sin(6*x), label = 'sin(6x)', color = 'purple', linestyle = 'dashdot')

ax.set(xlabel='x', ylabel='f(x)', title='A multiple plot')

ax.legend(loc='lower left')

ax.set_xlim(0,3*np.pi)
ax.set_ylim(-3,3)

ax.grid()

fig.savefig("multiple_test.png")
plt.show()
```

```python
# Multiple Subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

fig.suptitle('Multiple Subplots')
x = np.linspace(0, 3*np.pi, 1000)

axs[0,0].plot(x,np.sin(x), color = 'blue')
axs[0,0].set_title('$sin(x)$')

axs[0,1].plot(x,np.cos(x), color = 'red')
axs[0,1].set_title('cos(x)')

axs[1,0].plot(x,np.tan(x), color = 'green')
axs[1,0].set_title('tan(x)')

axs[1,1].plot(x,x**2, color = 'purple')
axs[1,1].set_title('$x^2$')
```
