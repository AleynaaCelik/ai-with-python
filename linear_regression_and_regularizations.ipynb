```python
!pip install mglearn
import mglearn
import numpy as np
import matplotlib.pyplot as plt
import sklearn
```

```python
X, Y = mglearn.datasets.make_wave(n_samples=400)
plt.plot(X, Y, 'o')
plt.xlabel('Feature')
plt.ylabel('Target')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state=0)
print(X_train.shape)
print(X_test.shape)
```

```python
# y ~ b0 + b1 * x1
from sklearn.linear_model import LinearRegression
lr_model = LinearRegression().fit(X_train, Y_train)
print(lr_model.coef_)
print(lr_model.intercept_)
```

```python
print('R2 for train set =', lr_model.score(X_train, Y_train))
print('R2 for test set =', lr_model.score(X_test, Y_test))
```

```python
## Ridge regression
from sklearn.linear_model import Ridge
ridge_model = Ridge(alpha=0.5).fit(X_train, Y_train)
print(ridge_model.coef_)
print(ridge_model.intercept_)

print('R2 for train set =', ridge_model.score(X_train, Y_train))
print('R2 for test set =', ridge_model.score(X_test, Y_test))
```

```python
## LASSO (Least Absolute Selection and Shrinkage Operator) Regression
from sklearn.linear_model import Lasso
lasso_model = Lasso(alpha=0.01, max_iter=10000).fit(X_train, Y_train)
print(lasso_model.coef_)
print(lasso_model.intercept_)

print('R2 for train set =', lasso_model.score(X_train, Y_train))
print('R2 for test set =', lasso_model.score(X_test, Y_test))
```
