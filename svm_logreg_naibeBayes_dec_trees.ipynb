```python
from sklearn.datasets import make_blobs
X, Y = make_blobs(random_state=42)
```

```python
plt.scatter(X[:,0], X[:,1], c=Y, s=60)
```

```python
from sklearn.svm import LinearSVC
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state=0)
```

```python
svc_model = LinearSVC().fit(X_train, Y_train)
```

```python
print(svc_model.coef_, svc_model.intercept_)
```

```python
test_accuracy = svc_model.score(X_test,Y_test)
train_accuracy = svc_model.score(X_train,Y_train)
```

```python
print(train_accuracy, test_accuracy)
```

```python
from sklearn.datasets import make_moons
X, Y = make_moons(random_state=42)
```

```python
plt.scatter(X[:,0], X[:,1], c=Y, s=60)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state=0)

for c in [0.0001, 0.001, 0.01, 0.1, 1, 10]:
  svc_model = LinearSVC(penalty='l2', C=c).fit(X_train, Y_train)
  test_accuracy = svc_model.score(X_test,Y_test)
  train_accuracy = svc_model.score(X_train,Y_train)
  print(train_accuracy, test_accuracy)
```
